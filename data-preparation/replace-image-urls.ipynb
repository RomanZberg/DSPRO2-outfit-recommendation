```python
import os

not_existent_images = 0

image_paths = []

for index, record in df.iterrows():
    path_to_img_folder = f'../datasets/raw/'
    img_folder_path = f'images/{str(record["set_id"])}/{str(record["clothingItemIndex"])}.jpg'

    full_path = f'{path_to_img_folder}/{img_folder_path}'

    if not os.path.exists(full_path):
        not_existent_images += 1
        print(f'path deos not exist "{full_path}"')
        image_paths.append(None)
    else:
        image_paths.append(img_folder_path)

print(f'not existent images: {str(not_existent_images)}')
```

```python
df = pd.concat([df, pd.Series(image_paths, name='imagePath')], axis=1)
```
